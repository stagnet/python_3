```python
df=pandas.read_json("supermarkets.json")# we chose the dataframe as for our next eg....
df
```

```python
df["Address"] = df["Address"]+ ", "+df["City"]+ ", "+df["State"]
df
```

```python
df['coordinate'] = df['Address'].apply(nomi.geocode)#create a new row 'coordinates' have usual addresses with thier respective coordinates...
df
```

```python
df.coordinate[1]
```

```python
df.coordinate[2]
```

```python
df['latitude'] = df['coordinate'].apply(lambda x:x.latitude if x!= None else None)# use lambda function to create a temprory variable which stores the logi & lati. value from df[coordinate]
df['longitude'] = df['coordinate'].apply(lambda x:x.longitude if x!= None else None)
```
